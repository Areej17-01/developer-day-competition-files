```python
tdata=pd.read_csv("train.csv")
df=tdata.copy()
test_df=pd.read_csv("test.csv")
dataset=pd.concat([df,test_df])
```

```python
tdata
```

```python
ds=dataset.copy()
for i in ds.columns:
    if ds[i].dtype=="O":
        ds[i] = pd.Categorical(ds[i]).codes
```

```python
X=ds.drop(columns=["mutation"])
#X=ds[sp]
y=ds['mutation']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train = X[:180000]
y_train = y[:180000]
```

```python
X_test=X[180000:]
X_test.shape
```

```python

y_test=y[180000:]
```

```python
cb= CatBoostClassifier()
cb.fit(X_train,y_train)
```

```python
pred=cb.predict(X_test)
pred=pd.DataFrame(pred)
test=pd.read_csv("test.csv")
y_pred=pd.concat([test["ID"],pred],axis=1)
y_pred.columns=["ID","mutation"]

y_pred.mutation=y_pred.mutation.astype(int)
```

```python
y_pred
```

```python
y_pred.to_csv('cb.csv', index = False)
```
